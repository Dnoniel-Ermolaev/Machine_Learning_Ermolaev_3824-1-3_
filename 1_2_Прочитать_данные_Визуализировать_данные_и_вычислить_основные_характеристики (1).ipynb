Преобразуем **"Budget"** в численный признак

```python
# Функция для преобразования строки бюджета в числовой формат
def convert_budget(budget_str):
    # Проверка на NaN (пустые значения)
    if pd.isna(budget_str):
        return None

    # Удаление символов "$" и ","
    budget_str = budget_str.replace('$', '').replace(',', '')

    # Преобразование в числовой формат
    try:
        budget = int(budget_str)
        return budget
    except ValueError:
        return None  # Если не удалось преобразовать, вернуть None

# Применение функции к столбцу "Бюджет" и перезапись этого признака к численному.
dataset['Budget'] = dataset['Budget'].apply(convert_budget)

# Вывод первых нескольких строк с новым числовым признаком
print("Пример первых нескольких строк с новым числовым признаком 'Budget':")
print(dataset[['Budget']].head())
```

Преобразуем **"Release Date"** в численный признак

Обратим внимание, что некоторые даты релиза представлены только в формате года, поэтому дабы упростить жизнь, преобразуем все даты только к году.

```python
# Преобразование "Release Date" в числовой признак
dataset['Release Date'] = pd.to_datetime(dataset['Release Date'], errors='coerce')

# Функция для извлечения года из даты или строки
def extract_year(date):
    if isinstance(date, pd.Timestamp):
        return date.year
    elif isinstance(date, pd.NaTType):
        return pd.NaT
    elif isinstance(date, str) and len(date) == 4 and date.isdigit():
        return int(date)
    else:
        return pd.NaT

dataset['Release Date'] = dataset['Release Date'].apply(extract_year)
```

Теперь надо решить вопрос с тем, как преобразовать такие признаки, как  **Filming Locations**, **Language**, **Cast**, **Genres**.
Рассмотрим в качестве примера признак **Genres**.

Очевидно, что один фильм может относится к разным жанрам, поэтому в столбцах часто можно встретить конструкцию "Comedy| Horror| Thriller". Если каждому такому примеру задавать конкретный номер, то мы получим очень большое количество разнообразных чисел, которые на самом деле будут обозначать практически одно и то же.

Поэтому разделим столбец жанров на несколько новых столбцов и добавим несколько новых бинарных признаков, каждый из которых будет обозначать принадлежность фильма к конкретному жанру.

```python
# Заменить значения NaN на пустые строки
dataset['Genres'].fillna('', inplace=True)
# Разделение строк по разделителю, создание новых столбцов и присвоение имен жанрам
genres = dataset['Genres'].str.get_dummies(sep='|')


# Создать столбцы с оригинальными названиями жанров и присвоить значения 1 или 0
for genre in genres.columns:
    dataset[genre] = dataset['Genres'].apply(lambda x: 1 if genre in x else 0)

dataset.drop('Genres', axis=1, inplace=True)
```

Аналогично преобразуем остальные оставшиеся признаки к бинарным. Преобразуем **Filming Locations**, **Language**, **Cast**.

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Функция для извлечения страны из строки
def extract_country(location):
    if isinstance(location, str):
        location_split = location.split(',')
        if len(location_split) > 1:  # Проверка наличия страны
            return location_split[-1].strip()
    return np.nan  # Замена отсутствующих стран значением NaN

# Применение функции для извлечения стран и создание нового столбца 'Country'
dataset['Country'] = dataset['Filming Locations'].apply(extract_country)

# Заполнение отсутствующих значений в столбце 'Country' перед кодированием
dataset['Country'].fillna('Unknown', inplace=True)  # Заполнение пустых значений другим значением или NaN
label_encoder = LabelEncoder()

# Преобразование стран в числовой формат
dataset['Filming Locations'] = label_encoder.fit_transform(dataset['Country'])

# Удаление временного столбца 'Country'
dataset.drop('Country', axis=1, inplace=True)

non_zero_filming_locations = dataset[dataset['Filming Locations'] != 0]
print(non_zero_filming_locations['Filming Locations'])

dataset
```

```python
# Получаем список всех уникальных языков из столбца 'Language'
unique_languages = dataset['Language'].str.split('|', expand=True).stack().unique()

# Сортируем языки по частоте появления
language_count = dataset['Language'].str.split('|').explode().value_counts()
top_40_languages = language_count.head(40).index.tolist()

# Создаем бинарные признаки для топ-40 языков
for language in top_40_languages:
    dataset[language] = dataset['Language'].str.contains(language, na=False).astype(int)

# Удаляем столбец 'Language' и сохраняем только созданные бинарные признаки
dataset.drop('Language', axis=1, inplace=True)

dataset
```

### **Подсчитаем выбросы по признакам, что не являются бинарными.**

Выбросы (outliers) - это значения в наборе данных, которые существенно отличаются от остальных значений в этом наборе. Это могут быть значения, которые значительно выходят за пределы ожидаемого диапазона и не соответствуют остальным данным.

Выбросы могут возникать по разным причинам:

**Естественные аномалии:** Некоторые данные могут быть аномальными из-за естественной изменчивости или редкости в данных. Например, при обработке медицинских данных - выбросами могут быть редкие заболевания.

**Ошибки измерения:** Ошибки при сборе или записи данных могут привести к выбросам. Например, опечатки при вводе данных или технические ошибки в измерительном оборудовании.

**Необычные события:** Иногда выбросы могут представлять собой необычные события, которые не соответствуют типичному поведению данных.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Выбор числовых признаков (количественных)
numeric_columns = dataset.select_dtypes(include=['number']).columns
selected_numeric_columns = numeric_columns[:7]

# Создание нового DataFrame, содержащего только выбранные числовые признаки
selected_numeric_data = dataset[selected_numeric_columns]

# Определение выбросов
outliers_dict = {}
for feature in selected_numeric_data.columns:
    Q1 = dataset[feature].quantile(0.25)
    Q3 = dataset[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1 * IQR
    upper_bound = Q3 + 1 * IQR
    outliers = (dataset[feature] < lower_bound) | (dataset[feature] > upper_bound)
    outliers_dict[feature] = outliers.sum()

# Вывод информации о выбросах
print("\nКоличество выбросов по каждому признаку:")
for feature, count in outliers_dict.items():
    print(f"{feature}: {count}")
```

Формула для вычисления среднего значения:
