**Датасет:**
https://www.kaggle.com/datasets/PromptCloudHQ/imdb-horror-movie-dataset/data

**Контекст:** Данный датасет содержит в себе более 3300 фильмов оценённых IMDB на момент 2012 года.

**Контент:**
1.  **Title** - Название фильма (Это ***текстовый*** признак, т.к. его нельзя категоризовать.)
2.  **Genres** - Жанр фильма (Это ***категориальный*** признак, т.к. к одному фильму могут относиться несколько жанров одновременно и их можно просто пронумеровать.)
3.  **Release Date** - Дата релиза фильма (В датасете он представлен в виде текста, но на деле его можно превратить как в ***категориальный*** признак, так и во ***временной***, зависит от кодирования.)
4.  **Release Country** - Страна релиза фильма (Это ***географический*** признак, однако я считаю, что его можно тоже рассматривать как ***категориальный*** ибо каждой уникальной стране можно присвоить номер.)
5.  **Movie Rating** - Рейтинг фильма, его возрастное ограничение ( Данный признак обладает наибольшим количеством пробелов. Если преобразовать текстовые названия в цифровые, то он тоже станет ***категориальным***.)
6.  **Review Rating** - Рейтинг фильма или другими словами, его оценка. (Это ***численный*** признак, из которого может получится ***категориальный*** признак или даже ***порядковый***, зависит от того, что мы планируем делать.)
7.  **Movie Run Time** - Это продолжительность фильма в минутах. (Это ***численный*** признак, или ***временной***, в зависимости от того как мы его преобразуем из текстового.)
8.  **Plot** - Это Сюжет и подробности фильма (Это ***текстовый*** признак.)
9.  **Cast** - Актерский состав фильма (Это ***категориальный*** признак ибо каждого нового встреченного актёра можно условно пронумеровать.)
10. **Language** - Язык фильма (Это тоже ***категориальный*** признак, т.к. каждому новому языку мы присваиваем свой порядковый номер.)
11. **Filming Locations** - Съёмочная локация (Данный признак представляет собой текстовое название страны, его можно превратить как в ***географический***, так и в ***категориальный*** признак.)
12. **Budget** - Бюджет фильма (Данный признак является ***численным*** признаком)


**Постановка задачи:**

Данный датасет можно использовать как для задачи классификации, так и для задачи регрессии.

Чтобы использовать данный датасет для задачи классификации, необходимо создать отдельный новый признак, который будет на основе оценке фильма определять тот к конкретному классу. В нашем случае будем отсекать новый класс по границам Review Rating [1, 3, 5, 6, 8, 9.8].

**Я же планирую решать задачу регрессии, дабы на основе представленных значений можно было предсказать рейтинг у фильма.**
