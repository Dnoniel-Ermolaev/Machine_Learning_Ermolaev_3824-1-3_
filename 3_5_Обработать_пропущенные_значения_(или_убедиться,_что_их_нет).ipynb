```python
# Выбор числовых столбцов
numeric_columns = dataset.select_dtypes(include=['int64', 'float64']).columns

# Интерполяция пропущенных значений в числовых столбцах
dataset[numeric_columns] = dataset[numeric_columns].interpolate(method='linear')

num_rows, num_columns = dataset.shape
print(f"Количество признаков в датасете: {num_columns}")
print(f"Количество объектов в датасете: {num_rows}")

# Заполнение пропущенных значений в числовых столбцах средним значением
for column in numeric_columns:
    mean_value = dataset[column].mean()
    dataset[column].fillna(mean_value, inplace=True)

# Получение типов данных признаков
data_types = dataset.dtypes

# Вывод типов данных
print("Типы данных признаков в датасете:")
print(data_types)

dataset
```

Для нормализации числовых признаков в датасете можно воспользоваться методами нормализации, такими как Min-Max Scaling или Z-нормализацией (стандартизацией). Здесь я покажу пример Z-нормализации с использованием библиотеки Pandas.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Выбор конкретных числовых столбцов для нормализации
columns_to_normalize = ['Release Date','Release Country', 'Movie Run Time','Movie Rating', 'Cast','Budget','Filming Locations']
# Создание экземпляра StandardScaler
scaler = MinMaxScaler()

# Нормализация конкретных числовых признаков
dataset[columns_to_normalize] = scaler.fit_transform(dataset[columns_to_normalize])

dataset
```
